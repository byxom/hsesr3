```python
import pandas as pd
```

Посмотрим, с какими данными мы имеем дело

```python
dat = pd.read_csv('data-final.csv',sep='\t')
pd.options.display.max_columns = 200
dat.head ()
```

И так, ну пофану можно попробовать предсказать страну проживания человека по его ответам в психологическом тесте

```python
from ludwig.api import LudwigModel
```

Создаю словарь входящих фич - а именно отбераю из всей таблицы только столбики с ответами на тесты

```python
input_features_n=list(dat.columns[:-60])
input_features=[]
for i in input_features_n:
    input_features.append({"name": i, "type": "numerical"})
```

создаю определение модели, пускай предсказывает страну

```python

model_definition = {
    "input_features": input_features,
    "output_features": [
                       {"name": "country", "type": "category"}
    ]
}
```

создаю объект модели

```python
model = LudwigModel(model_definition)
```

обучаю модель, делаю предсказания

```python
train = model.train(dat)
predictions = model.predict(dat)
```

Смотрю на кривые обучения, которые, честно сказать совсем не радуют, предсказания у нас получились практически бесполезные

```python
from ludwig.visualize import learning_curves
learning_curves(train, 'country')
```
